```python
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv('/kaggle/input/all-movies/All_Movies1.csv',lineterminator='\n')
df.head()
```

```python
# Calculating Weighted Average for each Movie Rating
R = df['vote_average']
v = df['vote_count']
C = df['vote_average'].mean()
m = df['vote_count'].quantile(0.85)
```

```python
df['Weighted Average'] = ((R*v) + (C*m))/(v+m)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df[['normalized_weighted_avg','normalized_popularity']] = scaler.fit_transform(df[['Weighted Average','popularity']])
```

```python
df['Score'] = df['normalized_weighted_avg'] * 0.8 + df['normalized_popularity'] * 0.2
```

```python
df = df[['Title','genres','poster_path','Score']]
df 
```

```python
# finding out different genres 
s = set()
for i in tqdm(df['genres']):
    for j in i.split(','):
        s.add(j)
```

```python
print(s)
```

```python
# total number of genres
len(s)
```

```python
for i in s:
    df.loc[:,i] = 0
```

```python
df
```

```python
for i in tqdm(df.index):
    for j in df.loc[i,'genres'].split(','):
        df.loc[i,j]=1
```

```python
df
```

```python
df = df.sort_values('Score',ascending=False)
```

```python
df.to_csv('Genre_DataFrame.csv')
```

```python
# Function to get top 10 movies with its poster_path according to specified genre
def get_movies(genre,n=10):
    temp = df[df[genre]==1][['Title','poster_path']].reset_index(drop=True)
    
    ans = {}
    for i in range(n): 
        ans[temp.loc[i,'Title']]=temp.loc[i,'poster_path']
    
    return ans
```
